```python
import polars as pl
import altair as alt
import json
```

# Load Data

```python
data = json.load(open("../out.json"))
data
```

```python
results = pl.json_normalize(data["results"]).with_columns(
    name = pl.col("name").str.replace("HashSet<usize, core::hash::BuildHasherDefault<fxhash::FxHasher>>", "FxHashSet<usize>", literal=True),
).with_columns(
    kind = pl.col("name").str.split("::").list.get(5),
    bench = pl.col("name").str.split("::").list.get(6),
    typ = pl.col("name").str.split("::").list.get(7),
    size = pl.col("name").str.split("::").list.get(8).cast(pl.Int64),
    time = pl.col("time.median_ns"),
    alloc_count = pl.col("alloc.alloc.count.median"),
    alloc_size = pl.col("alloc.alloc.size.median"),
)
results
```

## Plot Helper

```python
def plot_size_stats(df, y, scale_type = "linear", x_scale_type=None, y_scale_type=None):
    x_scale_type = alt.ScaleType(x_scale_type) if x_scale_type else alt.ScaleType(scale_type)
    y_scale_type = alt.ScaleType(y_scale_type) if y_scale_type else alt.ScaleType(scale_type)
    selection = alt.selection_point(fields=['typ'], bind='legend')
    return (
        df
        .filter((pl.col("kind") == "random"))
        .plot
        .line(
            x=alt.X('size:Q', scale=alt.Scale(type=x_scale_type, base=10)),
            y=alt.Y(f'{y}:Q', scale=alt.Scale(type=y_scale_type, base=10)),
            color='typ:N',
            opacity=alt.condition(selection, alt.value(1), alt.value(0.2))
        ).add_params(selection)
        .properties(height=200, width=500)
        .facet("bench", columns=2, )
        .resolve_scale(y='independent')
    )
```

# Time

```python
plot_size_stats(
    results
        .filter((pl.col("bench") != "bench_is_adjacent_large")),
    y="time",
)
```

```python
plot_size_stats(
    results
        .filter((pl.col("bench") != "bench_is_adjacent_large")),
    y="time",
    scale_type="log"
)
```

single is_adjacent: `SortedVector` is faster than `FxHash` up to about 400 items.
Otherwise `FxHash` is almost always the best choice.
Unsorted vectors are beneficial only for fast construction when only very few
lookups are needed for small (up to few dozen items).

```python
plot_size_stats(
    results
        .filter((pl.col("bench") == "bench_is_adjacent_large")),
    y="time",
    x_scale_type="log",
    y_scale_type="log",
)
```

# Allocations

## Allocation Sizes

```python
plot_size_stats(
    results
        .filter((pl.col("bench") != "bench_is_adjacent_large"))
        .filter((pl.col("bench") != "bench_is_adjacent")),
    y="alloc_size",
)
```

```python
plot_size_stats(
    results
        .filter((pl.col("bench") != "bench_is_adjacent_large"))
        .filter((pl.col("bench") != "bench_is_adjacent")),
    y="alloc_size",
    scale_type="log",
)
```

## Allocations counts

```python
plot_size_stats(
    results
        .filter((pl.col("bench") != "bench_is_adjacent_large"))
        .filter((pl.col("bench") != "bench_is_adjacent")),
    y="alloc_count",
)
```

`BTreeSet` allocates often and a lot of memory.
`FxHashSet` and `HashSet` allocates a the most memory, but does only a small
number of allocations.

```python
plot_size_stats(
    results
        .filter((pl.col("bench") != "bench_is_adjacent_large"))
        .filter((pl.col("bench") != "bench_is_adjacent"))
        .filter((pl.col("typ") != "BTreeSet<usize>")),
    y="alloc_count",
    y_scale_type="linear"
)
```
